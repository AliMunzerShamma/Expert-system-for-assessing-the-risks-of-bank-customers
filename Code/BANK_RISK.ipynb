# KnowledgeEngine Class -.-

```python
class Bank_Risk(KnowledgeEngine):
    res_Reputation = 0
    res_Financial_or_Legal = 0
    res_Operational = 0
    res_Adherence = 0
    res_Controls_Systems = 0
    res_Procedures = 0
    result = 0
    result_Control = 0
    Total_result = []
    ERFW = (int(RiskFactor_Reputation.get())+int(RiskFactor_Financial_Legal.get())+int(RiskFactor_Operational.get()))
    ECFW = (int(ControlFactors_Adherence.get())+int(ControlFactors_ControlsSystems.get())+int(ControlFactors_Procedures.get()))
    index=0
    strr=''
    #Total Risks_______________________________________________________________________________________________________________
    @Rule(Total_Risk_Calculate('Total_Risk_Calculate'))
    def Total_Risk_Calculate(Self):
        temp=np.around(((Bank_Risk.result + (2*(100-(Bank_Risk.result_Control))))/3))
        Bank_Risk.Total_result.append(temp)                          
        Self.declare(Rate_Risk(x=Bank_Risk.Total_result[Bank_Risk.index]),Rate_Risk(x=Bank_Risk.Total_result[Bank_Risk.index]))                
        print('Total Risk Is : ')
        print(Bank_Risk.Total_result[Bank_Risk.index])
        Bank_Risk.index=Bank_Risk.index+1
        
#_________________________________________________________________________________________________________________________________
    #Potential Risks_______________________________________________________________________________________________________________
    #Calculate The Potential Risks
    @Rule(Potential_risks_Calculate('Potential_risks_Calculate'))
    def Potential_risks_Calculate(Self):
        Bank_Risk.result=(Bank_Risk.res_Operational+Bank_Risk.res_Financial_or_Legal+Bank_Risk.res_Reputation)/Bank_Risk.ERFW
    #High
    @Rule(AND(Effect_of_risk('High_R'),Risk_factor('Reputation')))
    def Effect_of_risk_High_AND_Risk_factor_Reputation (Self):
        Bank_Risk.res_Reputation=int(RiskImpact_High.get())*int(RiskFactor_Reputation.get())
        
    @Rule(AND(Effect_of_risk('High_F'),Risk_factor('Financial_or_Legal')))
    def Effect_of_risk_High_AND_Risk_factor_Financial_or_Legal (Self):
        Bank_Risk.res_Financial_or_Legal=int(RiskImpact_High.get())*int(RiskFactor_Financial_Legal.get())
        
        
    @Rule(AND(Effect_of_risk('High_O'),Risk_factor('Operational')))
    def Effect_of_risk_High_AND_Risk_factor_Operational (Self):
        Bank_Risk.res_Operational=int(RiskImpact_High.get())*int(RiskFactor_Operational.get())
        
    #Med
    @Rule(AND(Effect_of_risk('Med_R'),Risk_factor('Reputation')))
    def Effect_of_risk_Med_AND_Risk_factor_Reputation (Self):
        Bank_Risk.res_Reputation=int(RiskImpact_Med.get())*int(RiskFactor_Reputation.get())
        
    @Rule(AND(Effect_of_risk('Med_F'),Risk_factor('Financial_or_Legal')))
    def Effect_of_risk_Med_AND_Risk_factor_Financial_or_Legal (Self):
        Bank_Risk.res_Financial_or_Legal=int(RiskImpact_Med.get())*int(RiskFactor_Financial_Legal.get())
        
    @Rule(AND(Effect_of_risk('Med_O'),Risk_factor('Operational')))
    def Effect_of_risk_Med_AND_Risk_factor_Operational (Self):
        Bank_Risk.res_Operational=int(RiskImpact_Med.get())*int(RiskFactor_Operational.get())
        
        
    #Low
    @Rule(AND(Effect_of_risk('Low_R'),Risk_factor('Reputation')))
    def Effect_of_risk_Low_AND_Risk_factor_Reputation (Self):
        Bank_Risk.res_Reputation=int(RiskImpact_Low.get())*int(RiskFactor_Reputation.get())
        
    @Rule(AND(Effect_of_risk('Low_F'),Risk_factor('Financial_or_Legal')))
    def Effect_of_risk_Low_AND_Risk_factor_Financial_or_Legal (Self):
        Bank_Risk.res_Financial_or_Legal=int(RiskImpact_Low.get())*int(RiskFactor_Financial_Legal.get())
        
    @Rule(AND(Effect_of_risk('Low_O'),Risk_factor('Operational')))
    def Effect_of_risk_Low_AND_Risk_factor_Operational (Self):
        Bank_Risk.res_Operational=int(RiskImpact_Low.get())*int(RiskFactor_Operational.get())
#_________________________________________________________________________________________________________________________________
#Control on the Potential Risks_______________________________________________________________________________________________________________
    #Calculate The Control Risks
    @Rule(Risk_control_Calculate('Risk_control_Calculate'))
    def Risk_control_Calculate(Self):
        Bank_Risk.result_Control = (Bank_Risk.res_Adherence+Bank_Risk.res_Controls_Systems+Bank_Risk.res_Procedures)/Bank_Risk.ECFW 
    #High
    @Rule(AND(Evaluation_of_control('High_A'),Control_factor('Adherence')))
    def Evaluation_of_control_High_AND_Control_factor_Adherence (Self):
        Bank_Risk.res_Adherence=int(ControlEvaluation_High.get())*int(ControlFactors_Adherence.get())
        
    @Rule(AND(Evaluation_of_control('High_C'),Control_factor('Controls_Systems')))
    def Evaluation_of_control_High_AND_Control_factor_Controls_Systems (Self):
        Bank_Risk.res_Controls_Systems=int(ControlEvaluation_High.get())*int(ControlFactors_ControlsSystems.get())
        
    @Rule(AND(Evaluation_of_control('High_P'),Control_factor('Procedures')))
    def Evaluation_of_control_High_AND_Control_factor_Procedures (Self):
        Bank_Risk.res_Procedures=int(ControlEvaluation_High.get())*int(ControlFactors_Procedures.get())
        
    #Med
    @Rule(AND(Evaluation_of_control('Med_A'),Control_factor('Adherence')))
    def Evaluation_of_control_Med_AND_Control_factor_Adherence (Self):
        Bank_Risk.res_Adherence=int(ControlEvaluation_Med.get())*int(ControlFactors_Adherence.get())
        
    @Rule(AND(Evaluation_of_control('Med_C'),Control_factor('Controls_Systems')))
    def Evaluation_of_control_Med_AND_Control_factor_Controls_Systems (Self):
        Bank_Risk.res_Controls_Systems=int(ControlEvaluation_Med.get())*int(ControlFactors_ControlsSystems.get())
        
    @Rule(AND(Evaluation_of_control('Med_P'),Control_factor('Procedures')))
    def Evaluation_of_control_Med_AND_Control_factor_Procedures (Self):
        Bank_Risk.res_Procedures=int(ControlEvaluation_Med.get())*int(ControlFactors_Procedures.get())
        
    #Low
    @Rule(AND(Evaluation_of_control('Low_A'),Control_factor('Adherence')))
    def Evaluation_of_control_Low_AND_Control_factor_Adherence (Self):
        Bank_Risk.res_Adherence=int(ControlEvaluation_Low.get())*int(ControlFactors_Adherence.get())
        
    @Rule(AND(Evaluation_of_control('Low_C'),Control_factor('Controls_Systems')))
    def Evaluation_of_control_Low_AND_Control_factor_Controls_Systems (Self):
        Bank_Risk.res_Controls_Systems=int(ControlEvaluation_Low.get())*int(ControlFactors_ControlsSystems.get())
        
    @Rule(AND(Evaluation_of_control('Low_P'),Control_factor('Procedures')))
    def Evaluation_of_control_Low_AND_Control_factor_Procedures (Self):
        Bank_Risk.res_Procedures=int(ControlEvaluation_Low.get())*int(ControlFactors_Procedures.get())
#________________________________________________________________________________________________________________________________
#                                                 the Beneficiary is not Account Owner Risk -.-
    
    @Rule(Beneficiary_Account_Owner('No'))
    def Beneficiary_Account_Owner_No(Self):
        print('High Risk')
        
# If the Beneficiary is Account Owner then he is High risk.

    @Rule(Beneficiary_Account_Owner('Yes'))
    def Beneficiary_Account_Owner_Yes(Self):
        print ('Beneficiary Account Owner Yes')
        Self.declare(CML_Inquiry(x=CML.get()))
#________________________________________________________________________________________________________________________________        
#                                                            CML Inquiry Risk -.-
    @Rule(CML_Inquiry(x=P(lambda x:x=='Yes')))
    def CML_Inquiry_Yes(Self):
        print('High Risk')
        
    @Rule(CML_Inquiry(x=P(lambda x:x=='No')))
    def CML_Inquiry_No(Self):
        Self.declare(Final_Risk("Final_Risk"))
        print('CML Inquiry No')
        Self.declare(Transactions(x = 1))       
        Self=Bank_Risk()
        Bank_Risk.reset(Self)
        print(nationality.get())
        Self.declare(Work_Nature(x=business_nature.get()))
        Bank_Risk.run(Self)
        #Here -.-
        Self=Bank_Risk()
        Bank_Risk.reset(Self)
        Self.declare(Nationality(x=nationality.get()))    
        Bank_Risk.run(Self)
#_________________________________________________________________________________________________________________________________        
#                                                            Nationality Risk -.-
    @Rule(Nationality(x=P(lambda x:x != 'AFGHANISTAN')),Nationality(x=P(lambda x:x != 'ALGERIA')),Nationality(x=P(lambda x:x != 'ARGENTINA')),Nationality(x=P(lambda x:x != 'BAHRAIN')),Nationality(x=P(lambda x:x != 'BRAZIL')),Nationality(x=P(lambda x:x != 'CHINA')),Nationality(x=P(lambda x:x != 'COLOMBIA')),Nationality(x=P(lambda x:x != 'CUBA')),Nationality(x=P(lambda x:x != 'DJIBOUTI')),Nationality(x=P(lambda x:x != 'EGYPT')),Nationality(x=P(lambda x:x != 'EQUATORIAL GUINEA')),Nationality(x=P(lambda x:x != 'GIBRALTAR')),Nationality(x=P(lambda x:x != 'GREECE')),Nationality(x=P(lambda x:x != 'INDIA')),Nationality(x=P(lambda x:x != 'INDONESIA')),Nationality(x=P(lambda x:x != 'IRAN')),Nationality(x=P(lambda x:x != 'IRAQ')),Nationality(x=P(lambda x:x != 'LEBANON')),Nationality(x=P(lambda x:x != 'KOREA')),Nationality(x=P(lambda x:x != 'KUWAIT')))    
    def Nationality_low(Self):
        if len(Bank_Risk.strr) >= 3 :
            Bank_Risk.strr=''
        Bank_Risk.strr = Bank_Risk.strr + '0'   
        print('Nationality Risk Is Low')
          
    @Rule(OR(Nationality(x=P(lambda x:x == 'AFGHANISTAN')),Nationality(x=P(lambda x:x == 'ALGERIA')),Nationality(x=P(lambda x:x == 'ARGENTINA')),Nationality(x=P(lambda x:x == 'BAHRAIN')),Nationality(x=P(lambda x:x == 'BRAZIL')),Nationality(x=P(lambda x:x == 'CHINA')),Nationality(x=P(lambda x:x == 'COLOMBIA')),Nationality(x=P(lambda x:x == 'CUBA')),Nationality(x=P(lambda x:x == 'DJIBOUTI')),Nationality(x=P(lambda x:x == 'EGYPT')),Nationality(x=P(lambda x:x == 'EQUATORIAL GUINEA')),Nationality(x=P(lambda x:x == 'GIBRALTAR')),Nationality(x=P(lambda x:x == 'GREECE')),Nationality(x=P(lambda x:x == 'INDIA')),Nationality(x=P(lambda x:x == 'INDONESIA')),Nationality(x=P(lambda x:x == 'IRAN')),Nationality(x=P(lambda x:x == 'IRAQ')),Nationality(x=P(lambda x:x == 'LEBANON')),Nationality(x=P(lambda x:x == 'KOREA')),Nationality(x=P(lambda x:x == 'KUWAIT'))))
    def Nationality_Other(Self):
        Self.declare(Total_Risk_Calculate('Total_Risk_Calculate'))
        
        Self.declare(Potential_risks_Calculate('Potential_risks_Calculate'))
        
        print('Calculate Nationality Potential Risk')
        Self.declare(Effect_of_risk('High_R'),Risk_factor('Reputation'))
        Self.declare(Effect_of_risk('Med_O'),Risk_factor('Operational'))        
        Self.declare(Effect_of_risk('High_F'),Risk_factor('Financial_or_Legal'))
        
        Self.declare(Risk_control_Calculate('Risk_control_Calculate'))
        
        print('Calculate Nationality Controle On The Potential Risk')
        Self.declare(Evaluation_of_control('High_A'),Control_factor('Adherence'))
        Self.declare(Evaluation_of_control('High_C'),Control_factor('Controls_Systems'))        
        Self.declare(Evaluation_of_control('High_P'),Control_factor('Procedures'))
        
        
#___________________________________________________________________________________________________________________________________
#                                                     Work Nature Risk -.-

    @Rule(OR(Work_Nature(x=P(lambda x:x == 'Lawyers')),Work_Nature(x=P(lambda x:x == 'Accountants')),Work_Nature(x=P(lambda x:x == 'Brokers who work for the benefit of the client'))))       
    def Lawyers_Accountants_Brokers_who_work_for_the_benefit_of_the_client (Self):        
        Self.declare(Total_Risk_Calculate('Total_Risk_Calculate'))
        
        Self.declare(Potential_risks_Calculate('Potential_risks_Calculate'))
        
        print('Calculate Work Nature Potential Risk 1')
        Self.declare(Effect_of_risk('Med_R'),Risk_factor('Reputation'))
        Self.declare(Effect_of_risk('Med_O'),Risk_factor('Operational'))        
        Self.declare(Effect_of_risk('Med_F'),Risk_factor('Financial_or_Legal'))
        
        Self.declare(Risk_control_Calculate('Risk_control_Calculate'))
        
        print('Calculate Work Nature Controle On The Potential Risk 1')
        Self.declare(Evaluation_of_control('High_A'),Control_factor('Adherence'))
        Self.declare(Evaluation_of_control('High_C'),Control_factor('Controls_Systems'))        
        Self.declare(Evaluation_of_control('Med_P'),Control_factor('Procedures'))
        
    @Rule(OR(Work_Nature(x=P(lambda x:x == 'Trading of real estate')),Work_Nature(x=P(lambda x:x == 'Trading of Jewelry')),Work_Nature(x=P(lambda x:x == 'Trading of cars'))))           
    def Trading_of_real_estate_Trading_of_Jewelry_Trading_of_cars (Self):        
        Self.declare(Total_Risk_Calculate('Total_Risk_Calculate'))
        
        Self.declare(Potential_risks_Calculate('Potential_risks_Calculate'))
        
        print('Calculate Work Nature Potential Risk 2')
        Self.declare(Effect_of_risk('Med_R'),Risk_factor('Reputation'))
        Self.declare(Effect_of_risk('High_O'),Risk_factor('Operational'))        
        Self.declare(Effect_of_risk('Med_F'),Risk_factor('Financial_or_Legal'))
        
        Self.declare(Risk_control_Calculate('Risk_control_Calculate'))
        
        print('Calculate Work Nature Controle On The Potential Risk 2')
        Self.declare(Evaluation_of_control('High_A'),Control_factor('Adherence'))
        Self.declare(Evaluation_of_control('High_C'),Control_factor('Controls_Systems'))        
        Self.declare(Evaluation_of_control('High_P'),Control_factor('Procedures'))
        
    @Rule(Work_Nature(x=P(lambda x:x != 'Trading of real estate')),Work_Nature(x=P(lambda x:x != 'Trading of Jewelry')),Work_Nature(x=P(lambda x:x != 'Trading of cars')),Work_Nature(x=P(lambda x:x != 'Lawyers')),Work_Nature(x=P(lambda x:x != 'Accountants')),Work_Nature(x=P(lambda x:x != 'Brokers who work for the benefit of the client')))    
    def Other_Work_Nature (Self):
        if len(Bank_Risk.strr) >= 3 :
            Bank_Risk.strr=''
        Bank_Risk.strr = Bank_Risk.strr + '0'   
        print('Work Nature Risk Is Low')
#___________________________________________________________________________________________________________________________________
#                                                  Transactions Risk -.-

    @Rule(Transactions(x=P(lambda x:x == -1)))
    def Transactions_Mony_High (Self):
        Self.declare(Total_Risk_Calculate('Total_Risk_Calculate'))
        
        Self.declare(Potential_risks_Calculate('Potential_risks_Calculate'))
        
        print('Calculate Transactions Potential Risk')
        Self.declare(Effect_of_risk('Med_R'),Risk_factor('Reputation'))
        Self.declare(Effect_of_risk('High_O'),Risk_factor('Operational'))        
        Self.declare(Effect_of_risk('High_F'),Risk_factor('Financial_or_Legal'))
        
        Self.declare(Risk_control_Calculate('Risk_control_Calculate'))
        
        print('Calculate Transactions Controle On The Potential Risk')
        Self.declare(Evaluation_of_control('High_A'),Control_factor('Adherence'))
        Self.declare(Evaluation_of_control('High_C'),Control_factor('Controls_Systems'))        
        Self.declare(Evaluation_of_control('High_P'),Control_factor('Procedures'))
      
    
            
    @Rule(Transactions(x=P(lambda x:x == 1)))
    def Transactions_Mony_Med(Self):
        boo=0
        for x1 in Customer_Informations[Full_Name.get()][0]:
            if(x1 > int(deposit_threshold.get())):
                boo = 1
        for x1 in Customer_Informations[Full_Name.get()][1]:
            if(x1 > int(Withdrawal_Threshold.get())):
                boo = 1
        if boo == 1:
            Self.declare(Transactions(x=-1))
        else:
            if len(Bank_Risk.strr) >= 3 :
                Bank_Risk.strr=''
            Bank_Risk.strr = Bank_Risk.strr + '0'   
            print('Transactions Risk Is Low')
#___________________________________________________________________________________________________________________________________
#                                                  Rate Risk -.-        
    @Rule (Rate_Risk(x=P(lambda x:x >=23 )) , Rate_Risk(x=P(lambda x:x <=35 )))
    def Rate_Risk_23_35(Self):
        if len(Bank_Risk.strr) >= 3 :
            Bank_Risk.strr=''
        Bank_Risk.strr = Bank_Risk.strr + '0'       
        
    @Rule (Rate_Risk(x=P(lambda x:x >=36 )) , Rate_Risk(x=P(lambda x:x <=55 )))
    def Rate_Risk_36_55(Self):
        if len(Bank_Risk.strr) >= 3 :
            Bank_Risk.strr=''
        Bank_Risk.strr = Bank_Risk.strr + '1'
    
    @Rule (Rate_Risk(x=P(lambda x:x >=56 )) , Rate_Risk(x=P(lambda x:x <=73 )))
    def Rate_Risk_56_73(Self):
        if len(Bank_Risk.strr) >= 3 :
            Bank_Risk.strr=''
        Bank_Risk.strr = Bank_Risk.strr + '2'
#___________________________________________________________________________________________________________________________________
#                                                  Final Risk -.-        
    @Rule(Final_Risk("Final_Risk"))
    def Final_Risk_High_Med_Low(Self):
        print("Final_Risk : ") 
        print(result_risk[Bank_Risk.strr])        
```
